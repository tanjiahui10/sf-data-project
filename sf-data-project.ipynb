```python

import numpy as np
import matplotlib.pyplot as plt  # To visualize
import pandas as pd  # To read data
from sklearn.linear_model import LinearRegression
```

```python
Year = decade.iloc[:, 0].values.reshape(-1, 1)  # values converts it into a numpy array
Cases = decade.iloc[:, 1].values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column
linear_regressor = LinearRegression()  # create object for the class
linear_regressor.fit(Year, Cases)  # perform linear regression
Cases_pred = linear_regressor.predict(Year)  # make predictions
```

```python
plt.scatter(Year, Cases)
plt.plot(Year, Cases_pred, color='red')
plt.show()
```

```python
plt.scatter(Year, Cases)
plt.plot(Year, Cases_pred, color='red')
plt.xlabel('Year')
plt.ylabel('Cases of Human or Animal Waste')
plt.title('Cases of Human or Animal Waste between 2009 - 2019')
plt.show()
```
